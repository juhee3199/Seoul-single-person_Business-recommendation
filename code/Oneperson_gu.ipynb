## 1인가구 데이터 불러오기

```python
hum1 = pd.read_csv('grp_dist.csv', na_values = '?', encoding='cp949') # 엑셀 파일 읽기 #encoding='utf-8'
hum1.head()
```

```python
#행정구역 수정한 내용으로 지역별 평균값 피벗테이블 만들기
pivot_hum1 = pd.pivot_table(hum1, index='자치구')  #지역별 평균값 #, aggfunc=np.average
pivot_hum1 = pivot_hum1['X1인세대']
pivot_hum1 = pd.DataFrame(pivot_hum1)
pivot_hum1.head()
```

### 1인가구수 대비 여가오락시설 

```python
shop_hum = pd.concat([pivot_hum1, shop_ent_pco, shop_ent_par, shop_ent_mov, shop_ent_yog, shop_eat_kor, shop_eat_duk, shop_eat_cof, shop_eat_fas ], axis=1) #, left_on='시군구명', right_on='자치구'
shop_hum.head()
```

```python
shop_hum.columns = ['1인가구수','PC/오락/당구/볼링등', '무도/유흥/가무', '연극/영화/극장', '요가/단전/마사지', '한식', '분식', '커피점/카페', '패스트푸드']
shop_hum.head()
```

```python
shop_hum['PC 비율'] = shop_hum['PC/오락/당구/볼링등'] / shop_hum['1인가구수']
shop_hum['무도 비율'] = shop_hum['무도/유흥/가무'] / shop_hum['1인가구수']
shop_hum['연극 비율'] = shop_hum['연극/영화/극장'] / shop_hum['1인가구수']
shop_hum['요가 비율'] = shop_hum['요가/단전/마사지'] / shop_hum['1인가구수']
shop_hum['한식 비율'] = shop_hum['한식'] / shop_hum['1인가구수']
shop_hum['분식 비율'] = shop_hum['분식'] / shop_hum['1인가구수']
shop_hum['커피점 비율'] = shop_hum['커피점/카페'] / shop_hum['1인가구수']
shop_hum['패스트푸드 비율'] = shop_hum['패스트푸드'] / shop_hum['1인가구수']


shop_hum
```
